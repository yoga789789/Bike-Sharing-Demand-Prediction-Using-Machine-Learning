### **Business Problem Understanding**

**Context**

Sistem bike-sharing merupakan bentuk modern dari penyewaan sepeda konvensional, di mana seluruh proses (mulai dari pendaftaran, peminjaman, hingga pengembalian) dilakukan secara otomatis. Pengguna dapat meminjam sepeda di satu lokasi dan mengembalikannya di lokasi lain dengan mudah. Saat ini, terdapat ratusan program bike-sharing di berbagai negara dengan total lebih dari lima ratus ribu sepeda yang beroperasi. Sistem ini semakin mendapat perhatian karena kontribusinya dalam mengurangi kemacetan, menjaga lingkungan, serta mendorong gaya hidup sehat.

Selain manfaatnya dalam kehidupan sehari-hari, sistem bike-sharing juga menghasilkan data yang sangat bernilai untuk keperluan penelitian. Tidak seperti moda transportasi lain seperti bus atau kereta, sistem ini mencatat secara detail durasi perjalanan serta lokasi awal dan akhir pengguna. Dengan karakteristik tersebut, bike-sharing dapat dipandang sebagai jaringan sensor virtual yang merekam pola mobilitas masyarakat perkotaan. Melalui analisis data ini, berbagai aktivitas dan peristiwa penting di suatu kota dapat diidentifikasi dan dipahami dengan lebih baik.

**Problem Statement**

Sistem bike-sharing seperti Capital Bikeshare menghasilkan data operasional yang sangat kaya, mencakup informasi waktu (tanggal dan jam), musim, kondisi cuaca, serta jumlah pengguna casual dan terdaftar. Data ini mencerminkan pola permintaan sepeda yang dinamis dan dipengaruhi oleh berbagai faktor yang saling berinteraksi secara kompleks.

Namun, permintaan sepeda tidak bersifat linier dan mengalami fluktuasi signifikan berdasarkan jam, hari libur, musim, serta kondisi cuaca. Pendekatan analisis tradisional atau deskriptif belum mampu menangkap hubungan kompleks antar variabel tersebut secara optimal. Akibatnya, perusahaan berisiko mengalami ketidakseimbangan antara permintaan dan ketersediaan sepeda, yang dapat menurunkan efisiensi operasional dan kualitas layanan.

Oleh karena itu, diperlukan pendekatan machine learning untuk memodelkan pola permintaan sepeda secara lebih akurat dengan memanfaatkan data historis. Model prediktif diharapkan mampu membantu perusahaan dalam memahami faktor-faktor utama yang memengaruhi jumlah peminjaman sepeda serta mendukung pengambilan keputusan berbasis data.

**Goals**

Berdasarkan permasalahan tersebut, sistem bike-sharing seperti Capital Bikeshare perlu memiliki tool berbasis machine learning yang mampu memprediksi jumlah permintaan sepeda pada waktu dan kondisi tertentu. Dengan memanfaatkan perbedaan pada berbagai fitur yang tersedia dalam data seperti jam penggunaan, musim, hari libur, serta kondisi cuaca, model prediktif diharapkan dapat memberikan estimasi permintaan sepeda yang lebih akurat. Prediksi ini dapat membantu perusahaan dalam mengelola ketersediaan sepeda secara lebih optimal, sehingga risiko stasiun yang kosong atau terlalu penuh dapat diminimalkan dan pengalaman pengguna tetap terjaga.

Bagi perusahaan bike-sharing, keberadaan prediction tool yang andal tidak hanya berperan dalam meningkatkan efisiensi operasional, tetapi juga mendukung pengambilan keputusan strategis jangka panjang. Pengelolaan armada yang lebih tepat sasaran dapat meningkatkan tingkat pemanfaatan sepeda, kepuasan pengguna, serta keandalan layanan. Dengan prediksi berbasis data, diharapkan sepeda-sepeda dapat ditempatkan di stasiun yang tepat pada jam tertentu sehingga tidak terjadi oversupply atau undersupply. Pada jam-jam sepi pengguna, perusahaan dapat memanfaatkan kesempatan untuk maintenance armadanya atau memindahkan (realokasi) sepeda ke stasiun yang lebih ramai. Pada akhirnya, peningkatan kualitas layanan tersebut berpotensi mendorong pertumbuhan jumlah pengguna dan frekuensi peminjaman, yang berdampak positif terhadap income, keberlanjutan, dan kinerja bisnis perusahaan.

**Analytic Approach**

Untuk menjawab permasalahan yang ada, langkah pertama yang dilakukan adalah menganalisis data penggunaan bike-sharing guna menemukan pola dari berbagai fitur yang tersedia, seperti waktu, musim, hari libur, dan kondisi cuaca, yang memengaruhi jumlah peminjaman sepeda. Analisis ini bertujuan untuk memahami karakteristik permintaan sepeda pada kondisi dan periode yang berbeda.

Selanjutnya, akan dibangun sebuah model regresi berbasis machine learning yang mampu memprediksi jumlah peminjaman sepeda pada waktu tertentu. Model ini diharapkan dapat menjadi tool prediksi permintaan yang membantu perusahaan bike-sharing dalam mengelola ketersediaan sepeda secara lebih optimal. Dengan adanya prediksi yang akurat, perusahaan dapat meningkatkan efisiensi operasional, menjaga kualitas layanan, serta mendukung pengambilan keputusan berbasis data.

**Metric Evaluation**

Evaluasi performa model regresi dalam analisis ini dilakukan menggunakan beberapa metrik, yaitu Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), Mean Absolute Percentage Error (MAPE), serta koefisien determinasi (R²). RMSE mengukur rataan akar kuadrat dari error sehingga memberikan penalti yang lebih besar terhadap kesalahan prediksi yang ekstrem. MAE merepresentasikan rataan nilai absolut dari selisih antara nilai aktual dan nilai prediksi, sedangkan MAPE menunjukkan rataan persentase error yang dihasilkan oleh model.

Selain metrik error tersebut, R² digunakan untuk mengukur kemampuan model dalam menjelaskan variasi data target, yaitu proporsi variasi jumlah peminjaman sepeda yang dapat dijelaskan oleh fitur-fitur independen dalam model. Nilai R² berada pada rentang 0 hingga 1, di mana nilai yang semakin mendekati 1 menunjukkan kemampuan model yang semakin baik dalam merepresentasikan hubungan antara variabel prediktor dan target.

Secara keseluruhan, semakin kecil nilai RMSE, MAE, dan MAPE serta semakin besar nilai R², maka semakin baik kinerja model dalam memprediksi jumlah peminjaman sepeda berdasarkan fitur-fitur yang digunakan.

```python
# Cek Multikolinearitas

from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df_model[['temp', 'atemp', 'hum', 'weathersit', 'holiday', 'season', 'hr']]

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

calc_vif(X)
```

Hasil pengujian multikolinearitas menggunakan Variance Inflation Factor (VIF) menunjukkan bahwa variabel temp dan atemp memiliki nilai VIF yang sangat tinggi, mengindikasikan multikolinearitas yang kuat. Oleh karena itu, variabel atemp di-drop dari model untuk meningkatkan stabilitas estimasi koefisien regresi.

#### **Encoding**

```python
categorical_features = ['season', 'weathersit', 'day', 'month']
numerical_features = ['temp', 'hum', 'hr', 'year']
binary_features = ['holiday']

transformer = ColumnTransformer(
    [
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), categorical_features),
        ('num', StandardScaler(), numerical_features),
        ('bin', 'passthrough', binary_features)
    ]
)

```

StandardScaler digunakan untuk menyamakan skala fitur numerik agar model—terutama yang berbasis jarak dan optimasi gradien—dapat belajar secara lebih stabil dan adil tanpa bias terhadap fitur berskala besar.

One-Hot Encoding digunakan karena beberapa fitur bersifat kategorikal dan tidak memiliki urutan numerik. Metode ini mencegah model salah menafsirkan adanya hubungan urutan antar kategori.

### **Modeling**

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )
    print(model_cv_mape, i)
    
    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())

    # R2
    model_cv_r2 = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='r2', 
        error_score='raise'
    )
    print(model_cv_r2, i)

    score_r2.append(model_cv_r2)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())
```

Berdasarkan evaluasi menggunakan cross-validation 5-fold dengan metrik RMSE, MAE, MAPE, dan R², XGBoost Regressor menunjukkan performa terbaik dibandingkan model lainnya. Model ini menghasilkan nilai error terendah serta nilai R² tertinggi, yang mengindikasikan kemampuan yang sangat baik dalam memprediksi jumlah peminjaman sepeda serta menjelaskan variasi data.

Random Forest Regressor menempati posisi kedua dengan performa yang sedikit lebih rendah dibandingkan XGBoost, namun memiliki tingkat stabilitas yang sangat baik ditunjukkan oleh nilai standar deviasi error yang paling kecil.

Sementara itu, Linear Regression dan KNN Regressor menunjukkan performa yang jauh lebih rendah, mengindikasikan adanya underfitting karena ketidakmampuan model linier dan berbasis jarak sederhana dalam menangkap pola non-linear dan musiman pada data.

**Hyperparameter Tuning**

Rujukan nilai dari parameter: https://xgboosting.com/random-search-xgboost-hyperparameters/

```python
from scipy.stats import uniform

# Kedalaman pohon
max_depth = [3, 5, 7, 9, 11]

# Learning rate
learning_rate = uniform(0.01, 0.29)

# Jumlah pohon
n_estimators = [200, 300, 400]

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = uniform(0.6, 0.4)

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = uniform(0.6, 0.4)

min_child_weight = [1, 3, 5, 7]

# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__regressor__max_depth': max_depth,
    'model__regressor__learning_rate': learning_rate,
    'model__regressor__n_estimators': n_estimators,
    'model__regressor__subsample': subsample,
    'model__regressor__min_child_weight': min_child_weight,
    'model__regressor__colsample_bytree': colsample_bytree
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = log_xgb

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=0)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 30,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error', 'r2'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 0  
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error', 'rank_test_r2']).head()
```
